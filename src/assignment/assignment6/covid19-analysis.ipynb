## Covid19 Analysis

```python
# import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')
import numpy as np
import seaborn as sns
```

```python
# read covid19_clean file and look at data snapshot
covid19_clean = pd.read_csv("covid_19_clean_complete.csv")
covid19_clean['Date'] = pd.to_datetime(covid19_clean['Date'], infer_datetime_format=True)
covid19_clean.head()
```

```python
# show statistics
covid19_clean.info()
```

We can see there are missing values in column Province/State and WHO Region.

```python
# get max date
covid19_clean['Date'].max()
```

```python
# highest number of cases top 10 countries
# group by Country/Region, sum on Confirmed and sort in descending order to get countries with Top 10 cases
groupby_statedf = covid19_clean.loc[covid19_clean['Date']=='2020-06-03'][['Country/Region','Confirmed']].groupby(['Country/Region']).agg({'Confirmed':'sum'})\
                            .reset_index().rename(columns={'Confirmed':'Total Confirmed Cases'})\
                            .sort_values('Total Confirmed Cases',ascending=False).head(10)
# convert number of cases in millions
groupby_statedf['Total Confirmed Cases'] = (groupby_statedf['Total Confirmed Cases'] / 1000000).round(2)
# groupby_statedf.head()
plt.figure(figsize=(12,8))
g=sns.barplot(groupby_statedf['Country/Region'].values, groupby_statedf['Total Confirmed Cases'].values)
plt.xlabel('Country', fontsize=12)
plt.xticks(rotation=90)
plt.ylabel('No. of Confirmed Cases(in millions)', fontsize=12)
plt.title("Top 10 countries with highest number of corona positive cases")
ax=g
#annotate axis = seaborn axis
for p in ax.patches:
             ax.annotate("%.2f" % p.get_height(), (p.get_x() + p.get_width() / 2., p.get_height()),
                 ha='center', va='center', fontsize=11, color='gray', xytext=(0, 20),
                 textcoords='offset points')
_ = g.set_ylim(0,2) #To make space for the annotations
plt.show()
```

We can see US has most number of confirmed positive cases as compared to other countries. Spain which is on second place has 1/6th of cases in US.

```python
# highest number of cases top 10 countries
# group by Country/Region, sum on Confirmed and sort in descending order to get countries with Top 10 cases
groupby_statedf_death = covid19_clean.loc[covid19_clean['Date']=='2020-06-03'][['Country/Region','Deaths']].groupby(['Country/Region'])\
                            .agg({'Deaths':'sum'})\
                            .reset_index().rename(columns={'Deaths':'Total Deaths'})\
                            .sort_values('Total Deaths',ascending=False).head(10)
# convert number of cases in thousands
groupby_statedf_death['Total Deaths'] = (groupby_statedf_death['Total Deaths'] / 1000).round(2)
# groupby_statedf_death.head()
plt.figure(figsize=(12,8))
g=sns.barplot(groupby_statedf_death['Country/Region'].values, groupby_statedf_death['Total Deaths'].values)
plt.xlabel('Country', fontsize=12)
plt.xticks(rotation=90)
plt.ylabel('No. of Deaths(in thousand)', fontsize=12)
plt.title("Top 10 countries with highest number of Deaths")
ax=g
#annotate axis = seaborn axis
for p in ax.patches:
             ax.annotate("%.2f" % p.get_height(), (p.get_x() + p.get_width() / 2., p.get_height()),
                 ha='center', va='center', fontsize=11, color='gray', xytext=(0, 20),
                 textcoords='offset points')
_ = g.set_ylim(0,120) #To make space for the annotations
plt.show()
```

```python
# highest number of cases top 10 countries
# group by Country/Region, sum on Confirmed and sort in descending order to get countries with Top 10 cases
groupby_statedf_recover = covid19_clean.loc[covid19_clean['Date']=='2020-06-03'][['Country/Region','Recovered']].groupby(['Country/Region'])\
                            .agg({'Recovered':'sum'})\
                            .reset_index().rename(columns={'Recovered':'Total Recovered'})\
                            .sort_values('Total Recovered',ascending=False).head(10)
# convert number of cases in thousands
groupby_statedf_recover['Total Recovered'] = (groupby_statedf_recover['Total Recovered'] / 1000).round(2)
# groupby_statedf_death.head()
plt.figure(figsize=(12,8))
g=sns.barplot(groupby_statedf_recover['Country/Region'].values, groupby_statedf_recover['Total Recovered'].values)
plt.xlabel('Country', fontsize=12)
plt.xticks(rotation=90)
plt.ylabel('No. of Recovered(in thousand)', fontsize=12)
plt.title("Top 10 countries with highest number of Recovered people")
ax=g
#annotate axis = seaborn axis
for p in ax.patches:
             ax.annotate("%.2f" % p.get_height(), (p.get_x() + p.get_width() / 2., p.get_height()),
                 ha='center', va='center', fontsize=11, color='gray', xytext=(0, 20),
                 textcoords='offset points')
_ = g.set_ylim(0,600) #To make space for the annotations
plt.show()
```

### Time Series Visualization

```python
# get aggregated data for all countries since there could be multiple entries for country with diff states
filtered_data = covid19_clean[['Country/Region','Date','Confirmed','Recovered','Deaths']]\
                        .groupby(['Country/Region','Date'])\
                        .agg({'Confirmed':'sum','Recovered':'sum','Deaths':'sum'})\
                        .reset_index()\
                        .sort_values(['Country/Region','Date'])
# get new cases per day by using difference with previous row since data is cumulative count 
filtered_data['New Cases/day'] = filtered_data['Confirmed'].diff()
filtered_data.loc[(filtered_data['New Cases/day'].isnull()==True) | \
                  (filtered_data['New Cases/day'] < 0) , 'New Cases/day']=0

# get number of new recovered patients per day by using difference with previous row since data is cumulative count 
filtered_data['New Recovered/day'] = filtered_data['Recovered'].diff()
filtered_data.loc[(filtered_data['New Recovered/day'].isnull()==True) | \
                  (filtered_data['New Recovered/day'] < 0) , 'New Recovered/day']=0

# get number of new deaths per day by using difference with previous row since data is cumulative count 
filtered_data['New Deaths/day'] = filtered_data['Deaths'].diff()
filtered_data.loc[(filtered_data['New Deaths/day'].isnull()==True) | \
                  (filtered_data['New Deaths/day'] < 0) , 'New Deaths/day']=0

# get data for last 3 months
filtered_data = filtered_data[(filtered_data['Date']>='2020-03-01') & (filtered_data['Date']<='2020-06-03')]\
                        .set_index('Date')
#filtered_data[filtered_data['Country/Region']=='US'].set_index('Date').head()
```

```python
# trend of new cases in top 5 countries in last 3 months
plt.figure(figsize=(12,12))
filtered_data.loc['2020-03':'2020-06', 'New Cases/day'][filtered_data['Country/Region']=='US'].plot(marker='o', linestyle='-')
filtered_data.loc['2020-03':'2020-06', 'New Cases/day'][filtered_data['Country/Region']=='Brazil'].plot(marker='o', linestyle='-')
filtered_data.loc['2020-03':'2020-06', 'New Cases/day'][filtered_data['Country/Region']=='Russia'].plot(marker='o', linestyle='-')
filtered_data.loc['2020-03':'2020-06', 'New Cases/day'][filtered_data['Country/Region']=='United Kingdom'].plot(marker='o', linestyle='-')
filtered_data.loc['2020-03':'2020-06', 'New Cases/day'][filtered_data['Country/Region']=='Spain'].plot(marker='o', linestyle='-')
plt.ylabel('Number of new cases per day', fontsize=12)
plt.xlabel("Date")
plt.legend(labels=['US', 'Brazil', 'Russia', 'United Kingdom', 'Spain'])
```

```python
# trend of new deaths in top 5 countries in last 3 months
plt.figure(figsize=(12,12))
filtered_data.loc['2020-03':'2020-06', 'New Deaths/day'][filtered_data['Country/Region']=='US'].plot(marker='o', linestyle='-')
filtered_data.loc['2020-03':'2020-06', 'New Deaths/day'][filtered_data['Country/Region']=='United Kingdom'].plot(marker='o', linestyle='-')
filtered_data.loc['2020-03':'2020-06', 'New Deaths/day'][filtered_data['Country/Region']=='Italy'].plot(marker='o', linestyle='-')
filtered_data.loc['2020-03':'2020-06', 'New Deaths/day'][filtered_data['Country/Region']=='Brazil'].plot(marker='o', linestyle='-')
filtered_data.loc['2020-03':'2020-06', 'New Deaths/day'][filtered_data['Country/Region']=='France'].plot(marker='o', linestyle='-')
plt.ylabel('Number of new deaths per day', fontsize=12)
plt.xlabel("Date")
plt.legend(labels=['US', 'UK', 'Italy', 'Brazil', 'France'])
```

```python
# trend of number of patients recovered in top 5 countries in last 3 months
plt.figure(figsize=(12,12))
filtered_data.loc['2020-03':'2020-06', 'New Recovered/day'][filtered_data['Country/Region']=='US'].plot(marker='o', linestyle='-')
filtered_data.loc['2020-03':'2020-06', 'New Recovered/day'][filtered_data['Country/Region']=='Brazil'].plot(marker='o', linestyle='-')
filtered_data.loc['2020-03':'2020-06', 'New Recovered/day'][filtered_data['Country/Region']=='Russia'].plot(marker='o', linestyle='-')
filtered_data.loc['2020-03':'2020-06', 'New Recovered/day'][filtered_data['Country/Region']=='Germany'].plot(marker='o', linestyle='-')
filtered_data.loc['2020-03':'2020-06', 'New Recovered/day'][filtered_data['Country/Region']=='Italy'].plot(marker='o', linestyle='-')
plt.ylabel('Number of new recoveries per day', fontsize=12)
plt.xlabel("Date")
plt.legend(labels=['US', 'Brazil', 'Russia', 'Germany', 'Italy'])
```
